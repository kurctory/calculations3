### Интегральное уравнение Фредгольма второго рода

Найти приближенное решение интегрального уравнения  
$u(x) + 0.1\int\limits^1_0sh(xy^2)u(y)dy = x + 0.1$  
используя квадратурную формулу - составную формулу трапеций

```python
import numpy as np
from math import sinh
import pandas as pd
```

```python
def H(x, y):
    return sinh(x * (y ** 2))

def kroneckerDelta(j, k):
    if j == k:
        return 1
    else:
        return 0
def f(x):
    return x + 0.1
```

```python
a, b = 0, 1
x = [a, (a + b) / 2, b]

def solve(n):
    D = np.zeros((n, n))
    g = np.zeros((n, 1))
    arr_x = np.linspace(a, b, n)

    A = np.zeros((n, 1))
    h = (b - a) / n
    for i in range(n):
        if i == 0 or n - 1:
            A[i] = h
        else:
            A[i] = 2 * h

    for j in range(n):
        g[j] = f(arr_x[j])
        for k in range(n):
            D[j][k] = kroneckerDelta(j, k) - A[k] * H(arr_x[j], arr_x[k])

    coeffs = np.linalg.solve(D, g)
    result = [0., 0., 0.]
    for i in range(3):
        tmp = 0
        for j in range(n):
            tmp += A[j] * H(x[i], arr_x[j]) * coeffs[j]
        result[i] = tmp + f(x[i])
    return result

```

```python
def norm(res1, res2):
    y = []
    for i in range(3):
        tmp = abs(res2[i] - res1[i])
        y.append(tmp)
    return max(y[0], y[1], y[2])

#n — начальное количество разбиений в составной формуле
n = 1
print("При начальном значении n = ", n)
data = []
delta = 1
while delta > 0.01:
    res_n = solve(n)
    res_2n = solve(2 * n)
    delta = norm(res_n, res_2n)
    data.append(res_n)
    n *= 2
    if delta <= 0.01:
        y = []
        for i in range(3):
            tmp = abs(res_2n[i] - res_n[i])
            y.append(tmp)

        data.append(res_2n)
        data.append(y)


columns = [
    "a",
    "(a + b) / 2",
    'b'
]
N = len(data)
indexes = [i*2 for i in range(1, N + 1)]
indexes[len(indexes) - 1] = 'u_2mn - u_mn'
table = pd.DataFrame(data, columns=columns, index=indexes)
table
```
